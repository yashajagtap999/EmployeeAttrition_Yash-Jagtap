## Task 4 — Model Building & Comparison


Split data into training and test sets (80/20)


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training Data Shape:", X_train.shape)
print("Testing Data Shape:", X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

Train the following 3 models:


```python
lr = LogisticRegression(
    class_weight='balanced',
    random_state=42,
    max_iter=1000
)

lr.fit(X_train, y_train)
```

```python
rf = RandomForestClassifier(
    n_estimators=100,
    class_weight='balanced',
    random_state=42
)

rf.fit(X_train, y_train)
```

```python
gb = GradientBoostingClassifier(
    random_state=42
)

gb.fit(X_train, y_train)
```

```python
lr_pred = lr.predict(X_test)
rf_pred = rf.predict(X_test)
gb_pred = gb.predict(X_test)
```

## Task 5 — Model Evaluation


```python
from sklearn.metrics import(
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    roc_auc_score,
    precision_score,
    recall_score,
    f1_score,
 accuracy_score
)

import matplotlib.pyplot as plt
import pandas as pd
```

```python
def evaluate_model(model, X_test, y_test):

    # Predictions
    y_pred = model.predict(X_test)

    # Probabilities for ROC-AUC
    y_prob = model.predict_proba(X_test)[:, 1]

    # Metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_prob)

    print(classification_report(y_test, y_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)

    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap='Blues')
    plt.show()

    return accuracy, precision, recall, f1, roc_auc
```

```python
lr_results = evaluate_model(lr, X_test, y_test)
```

```python
rf_results = evaluate_model(rf, X_test, y_test)
```

```python
gb_results = evaluate_model(gb, X_test, y_test)
```

```python
comparison = pd.DataFrame({
    "Model": [
        "Logistic Regression",
        "Random Forest",
        "Gradient Boosting"
    ],
    "Accuracy": [
        lr_results[0],
        rf_results[0],
        gb_results[0]
    ],
    "Precision": [
        lr_results[1],
        rf_results[1],
        gb_results[1]
    ],
    "Recall": [
        lr_results[2],
        rf_results[2],
        gb_results[2]
    ],
    "F1-Score": [
        lr_results[3],
        rf_results[3],
        gb_results[3]
    ],
    "ROC-AUC": [
        lr_results[4],
        rf_results[4],
        gb_results[4]
    ]
})

comparison = comparison.round(3)
comparison
```

```python
best_model = comparison.loc[comparison["ROC-AUC"].idxmax()]
print(best_model)
```

## Task 7 — HR Insights & Business Recommendations


**1. Which 3 factors most strongly predict that an employee will leave?**
According to the feature importance analysis, the three strongest predictors of employee attrition are **Monthly Income**, **Age**, and **Total Working Years**. Employees with lower salaries, younger employees, and those with fewer years of work experience are more likely to leave the company.

**2. Which department or job role should HR prioritize for retention efforts?**
Based on the exploratory data analysis, the **Sales Department**, particularly the **Sales Representative** job role, has the highest attrition rate. HR should prioritize retention programs for employees in these roles.

**3. Does salary alone explain attrition, or are there other stronger factors?**
No. Although **Monthly Income** is an important factor, salary alone does not explain employee attrition. Other factors such as **Age**, **Total Working Years**, **Years at Company**, **Distance from Home**, and **OverTime** also have a significant influence on whether an employee decides to leave.

**4. Write 2 concrete HR recommendations.**

* Introduce retention and career development programs for younger and early-career employees, including mentoring, training, and regular career discussions.
* Conduct periodic salary reviews and stay interviews for employees in the Sales department and other high-risk groups to address concerns before they decide to resign.

**5. What limitation does this model have?**
This model is based only on historical HR data and cannot capture personal reasons for resignation, such as family commitments, health issues, better external job opportunities, or career aspirations. Therefore, the model should be used as a decision-support tool rather than the sole basis for HR decisions.
